# NumPy Упражнения

Теперь, когда мы узнали о NumPy, давайте проверим ваши знания. Мы начнем с нескольких простых задач, а затем вам будут заданы более сложные вопросы.

#### Импорт NumPy как np

```python
import numpy as np
```

#### Создание массива из 10 нулей

```python
zeros = np.zeros(10)
print("Массив из 10 нулей:\n", zeros)
```

#### Создайте массив из 10 единиц

```python
ones = np.ones(10)
print("Массив из 10 единиц:\n", ones)
```

#### Создайте массив из 10 пятерок

```python
fives = np.full(10, 5)
print("Массив из 10 пятерок:\n", fives)
```

#### Создадим массив целых чисел от 10 до 50

```python
arr_10_to_50 = np.arange(10, 51)
print("Массив от 10 до 50:\n", arr_10_to_50)
```

#### Создадим массив из всех четных целых чисел от 10 до 50

```python
even_10_to_50 = np.arange(10, 51, 2)
print("Чётные числа от 10 до 50:\n", even_10_to_50)
```

#### Создание матрицы 3x3 со значениями в диапазоне от 0 до 8¶

```python
mat_3x3 = np.arange(9).reshape(3, 3)
print("Матрица 3x3 от 0 до 8:\n", mat_3x3)
```

#### Создание единичной матрицы 3x3

```python
identity_3x3 = np.eye(3)
print("Единичная матрица 3x3:\n", identity_3x3)
```

#### Используйте NumPy для генерации случайного числа от 0 до 1

```python
rand_num = np.random.rand()
print("Случайное число от 0 до 1:\n", rand_num)
```

#### Используйте NumPy для создания массива из 25 случайных чисел, отобранных из стандартного нормального распределения

```python
rand_normal = np.random.randn(25)
print("25 чисел из N(0,1):\n", rand_normal)
```

#### Создайте матрицу 10 на 10 из равномерно удаленных точек от 0.1 до 1( см linspace):

```python
mat_10x10 = np.linspace(0.1, 1, 100).reshape(10, 10)
print("Матрица 10x10 от 0.1 до 1:\n", mat_10x10)
```

#### Создайте массив из 20 линейно удаленных точек от 0 до 1( см linspace):

```python
linspace_20 = np.linspace(0, 1, 20)
print("20 точек от 0 до 1:\n", linspace_20)
```

#### Получить сумму всех значений в mat

```python
mat = mat_10x10
total_sum = np.sum(mat)
print("Сумма всех значений в mat:\n", total_sum)
```

#### Получить стандартное отклонение значений в mat (пакет Статистика)

```python
std_dev = np.std(mat)
print("Стандартное отклонение в mat:\n", std_dev)
```

#### Посчитайте сумму всех столбцов в mat

```python
sum_columns = np.sum(mat, axis=0)
print("Сумма по столбцам:\n", sum_columns)
```
